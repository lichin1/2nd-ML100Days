```python
# 載入套件 
# numpy 是資料處理用的套件
# matplotlib.pyplot 是畫圖用的套件
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = 3
b = 0.5
```

```python
# np.linspace(Start,End,n) n等分-From Start to End 
x_lin = np.linspace(0,100,101)
```

```python
# np.random.rand(n) 隨機生成 Standard Normal的 n 個亂數
y = (x_lin + np.random.randn(101) * 5) * w + b
```

```python
# plt.plot(x,y,'b.') 畫出 x,y 的線圖，'b.'=散佈圖 'r-' =迴歸線,
# plt.legend(loc=1) 畫出legend, 1=Upperright，2=Upperleft,3=Lowerleft,4=Lowerright 
# plt.title="" 顯示圖形標題
plt.plot(x_lin,y,'b.',label="data points")
plt.title("Assume we have data points")
plt.legend(loc=2)
plt.show()
```

```python
y_hat = x_lin * w + b
plt.plot(x_lin, y, 'b.', label = 'data')
plt.plot(x_lin, y_hat, 'r-', label = 'prediction')
plt.title("Assume we have data points (And the prediction)")
plt.legend(loc = 2)
plt.show()
```

```python
# def function_name(v1,v2,...): 定義函數，用 return 回傳要的值
# %3.f 把 % 之後的數值顯示到小數第三位數

def mean_absolute_error(y, yp):
    """
    計算 MAE
    Args:
        - y: 實際值
        - yp: 預測值
    Return:
        - mae: MAE
    """
    mae = MAE = sum(abs(y - yp)) / len(y)
    return mae

MAE = mean_absolute_error(y, y_hat)
print("The Mean absolute error is %.3f" % (MAE))
```
